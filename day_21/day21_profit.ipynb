```python
target=df['Profit']
feactures=df[['R&D Spend']]
```

```python
from sklearn.linear_model import LinearRegression
model_simple=LinearRegression().fit(feactures,target)
```

```python
model_simple.score{feactures,target}
```

```python
from sklearn.preprocessing import PolynomialFeatures as pf
```

```python
feat_poly1=pf(degree=1).fit_transform(feactures)
feat_poly2=pf(degree=2).fit_transform(feactures)
feat_poly3=pf(degree=3).fit_transform(feactures)
feat_poly4=pf(degree=4).fit_transform(feactures)
```

```python
print('Model Simple:',100*model_simple.score(feactures,target))

model_poly1=LinearRegression().fit(feat_poly1,target)
print('Model Poly1:',model_poly1.score(feat_poly1,target)*100)

model_poly2=LinearRegression().fit(feat_poly2,target)
print('Model Poly2:',model_poly2.score(feat_poly1,target)*100)

model_poly3=LinearRegression().fit(feat_poly3,target)
print('Model Poly3:',model_poly3.score(feat_poly1,target)*100)

model_poly4=LinearRegression().fit(feat_poly4,target)
print('Model Poly4:',model_poly4.score(feat_poly1,target)*100)
```
